```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
#house_price_dataset = sklearn.datasets.load_boston()
from sklearn.datasets import fetch_california_housing
house_price_dataset = fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns=house_price_dataset.feature_names)
```

```python
house_price_dataframe.head()
```

```python
# add the targets (price) column to the Dataframe

house_price_dataframe['Price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
house_price_dataframe.shape
```

```python
# check for missing values
house_price_dataframe.isnull().sum()
```

```python
house_price_dataframe.describe()
```

```python
# data correlation
correlation = house_price_dataframe.corr()
```

```python
# contructing a heatmap to understand the correlation
plt.figure(figsize=(10, 10))

sns.heatmap(correlation, cbar=True, square=True, fmt='.2f', annot=True, annot_kws={'size':8}, cmap='Reds')
```

```python
# split the data into data and labels/Target

X = house_price_dataframe.drop(['Price'], axis=1)
Y = house_price_dataframe['Price']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Split the data into training and test

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
# Model Training using XGBoost Regressor

model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
# Evaluation
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean absolute error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error: ", score_1)
print("Mean absolute error: ", score_2)
```

```python
# Visaulizing the actual prices and predicted prices

plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

```python
# Prediction on test data
test_data_prediction = model.predict(X_test)

print(test_data_prediction)

# R squarred error
score_3 = metrics.r2_score(Y_test, test_data_prediction)

# Mean absolute error
score_4 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error: ", score_3)
print("Mean absolute error: ", score_4)
```
